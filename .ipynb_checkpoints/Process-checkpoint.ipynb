```python
import pandas as pd
import pickle
import numpy as np
from collections import Counter
```

```python
# Disable warnings output
import warnings
warnings.filterwarnings("ignore")
```

```python
with open("./data/site.pkl", "rb") as sites:
    sites_mapping = pickle.load(sites)
```

```python
# Add empty site to mapping
sites_mapping["empty"] = 0
```

```python
train = pd.read_csv("./data/train.csv", index_col="session_id", parse_dates=[f"time{i}" for i in range(1, 11)])
test = pd.read_csv("./data/test.csv", index_col="session_id", parse_dates=[f"time{i}" for i in range(1, 11)])
```

```python
train
```

```python
# Order by date from oldest time1 to newsest time1
if not train.time1.is_monotonic_increasing: train = train.sort_values(by='time1', ascending=True)
if not test.time1.is_monotonic_increasing: test = test.sort_values(by='time1', ascending=True)
```

```python
# Create mapping of site columns and time columns
sites_columns = [f"site{i}" for i in range(1, 11)]
times_columns = [f"time{i}" for i in range(1, 11)]
```

```python
0 in list(sites_mapping.values()) # There is not site with id 0 in sites_mapping, so use it to fill na's
```

```python
# Fill n/a sites to 0 index
train[sites_columns] = train[sites_columns].fillna(0).astype('int')
test[sites_columns] = test[sites_columns].fillna(0).astype('int')
```

```python
# Replace site id to site domain string
sites_tmp = {v: k for k, v in sites_mapping.items()}
for site in sites_columns:
    train[site] = train[site].map(sites_tmp)
    test[site] = test[site].map(sites_tmp)
```

```python
# Delete duplicates
train = train.drop_duplicates()
```

```python
# Features based on time/date
def td_features(df):
    df_with_time = df[times_columns]
    lst_with_time_for_sites = []
    sessions_with_times = df[times_columns].values.tolist()
    df['morning_time'] = df['time1'].apply(lambda x: 1 if (x.hour >= 7) & (x.hour <= 11) else 0).astype('int') # Is session started at morning
    df['day_time'] = df['time1'].apply(lambda x: 1 if (x.hour >= 12) & (x.hour <= 18) else 0).astype('int') # Is session started at day time
    df['evening_time'] = df['time1'].apply(lambda x: 1 if (x.hour >= 19) & (x.hour <= 23) else 0).astype('int') # Is session started at evening
    df['night_time'] = df['time1'].apply(lambda x: 1 if (x.hour >= 0) & (x.hour <= 6) else 0).astype('int') # Is session started at night
    df['start_session'] = df['time1'].apply(lambda x: x.hour).astype('int') # Hour of session start (based on first site)
    df['day_of_week'] = df['time1'].apply(lambda x: x.dayofweek).astype('int') # Day of week of this session (based on first site)
    df['month'] = df['time1'].apply(lambda x: x.month).astype('int') # Month of session (based on first site)
    df['year'] = df['time1'].apply(lambda x: x.year).astype('int') # Year of session (based on first site)
    df['yearmonth'] = df['time1'].apply(lambda x: (100 * x.year + x.month)) # YEARMONTH of session (based on first site)
    df['session_duration'] = np.log1p((df_with_time.max(axis=1) - df_with_time.min(axis=1)).astype('timedelta64[s]').astype('int').values)

    # Get session start time type
    return df
```

```python
train = td_features(train)
test = td_features(test)
```

```python
repeated = Counter(train[sites_columns].values.flatten())

number_of_sites = len(set(train[sites_columns].values.flatten()))

def session_popularity(session):
    curr_site = 0
    for site in session:
        sites_index = (repeated[site]) / number_of_sites
    curr_site += sites_index
    return curr_site


def sites_features(df):
    df['sites_popularity_coeff'] = df[sites_columns].apply(lambda s: session_popularity(s), axis=1) # Coeff calculating for all known sites in session (avg)
    return df
```

```python
train = sites_features(train)
test = sites_features(test)
```

```python
train["is_short"] = train[sites_columns].apply(lambda x:1 if ((~np.isin(x, ['empty'])).sum() < 10) else 0, axis=1) # Low session time
test["is_short"] = test[sites_columns].apply(lambda x:1 if ((~np.isin(x, ['empty'])).sum() < 10) else 0, axis=1) # Low session time
```

```python
# Serizlize data with pickle
with open("./pickled/train.pkl", "wb") as fd:
    pickle.dump(train, fd)
with open("./pickled/test.pkl", "wb") as fd:
    pickle.dump(test, fd)
```
