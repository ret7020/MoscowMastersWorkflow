```python
import pandas as pd
import random
#df_test = pd.read_csv("test.csv")
#df_train = pd.read_csv("train.csv")


sessions = [[j] + [random.randint(0, 20) for i in range(10)] + [bool(random.randint(0, 1))] for j in range(2000)]
df_test = pd.DataFrame(sessions, columns = ["session_id", "site_0", "site_1", "site_2", "site_3", "site_4", "site_5", "site_6", "site_7", "site_8", "site_9", "target"])
df_test
```

```python
sns.pairplot( 
    data = df_test,
    hue = 'target',
    corner = True 
    ).fig.suptitle(
        '  ',
    y = 1.08
);
```

```python
sites_hits = []
for i in range(0, 10):
    sites_hits.append((i, df_test.loc[df_test['target'], f'site_{i}'].sum()))

site_to_target = pd.DataFrame(sites_hits, columns = ['ID', 'Target'])
site_to_target
```

```python
sns.lmplot(data=site_to_target, x="ID", y="Target")
```

```python
sns.barplot(site_to_target, x = "ID", y="Target")
```

```python
corr = df_test.groupby('target').sum().T.corr()
#  heatmap
sns.set(rc={'figure.figsize':(12, 12)})
sns.heatmap(corr, cmap='coolwarm', square=True, annot=True)
```
